```python
#Import packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
#read dataset

df = pd.read_csv(r"C:\Users\Samuel Oluwatoba\Documents\ZaranTech\Assignments 2\ML Project Hiring Challenge\ML Project Hiring Challenge\Hiring_Challenge.csv")
```

```python
#understand dataset

df.head()
df.isnull()
df.describe()
df.dtypes
```

```python
#transform strings labels to floats using LabelEncoder

from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
```

```python
#drop useless variables/columns that has no efect on our dependent variable, eg, C6
#and re-adjust dataset, obtaining our dependent and independent variables 

x= df.drop(["C6", "Hired"], axis = 1)
y = df.Hired
```

```python
#split dataset for train and test purposes

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.7)
```

```python
#Train/Biuld model
lm= LogisticRegression()
lm.fit(x_train, y_train)
```

```python
#make predictions based on the trained model
lm.predict(x_test)
```

```python
#Check accuracy of model
lm.score(x_test, y_test)
```

```python
#create a table to see how our model faired


df1 = pd.concat([x_test, y_test], axis=1)
df1.assign(prediction = lm.predict(x_test))
```
